## Introduction to Tensors

### Creating Tensors

PyTorch tensors are created using 'torch.Tensor ()' = https://pytorch.org/docs/stable/tensors.html

```python
# scalar == Um scalar (ou escalar, em português) é um único número, sem nenhuma dimensão.
scalar = torch.tensor(7)
scalar
```

```python
#Esse comando retorna o número de dimensões (ou “eixos”) de um tensor.
scalar.ndim
```

```python
#Get the tensor back as Python int
scalar.item()
```

```python
# Vector (possui tamanho e direção(x,y) = uma reta no plano)
vector = torch.tensor([7,7])
vector
```

```python
# ndim é um atributo que retorna o número de dimensões (ou eixos) de um array ou tensor.
vector.ndim
```

```python
# shape é um atributo que mostra o formato (ou tamanho em cada dimensão) de um array, vetor ou tensor.
vector.shape
```

```python
# Matrix --- criação de matriz
MATRIX = torch.tensor([[7,8],
                       [9,10]])
MATRIX
```

```python
MATRIX.ndim
```

```python
MATRIX[1]
```

```python
MATRIX.shape
```

```python
# TENSOR
TENSOR = torch.tensor([[[1,2,3],
                        [3,6,9],
                        [2,4,5]]])
TENSOR
```

```python
TENSOR.ndim
```

```python
TENSOR[0]
```

```python
TENSOR.shape
```

### Random tensors

Why Random tensors?

Random tensors are important because the way many neural networks learn is that they start with tensor full of random numbers and the adjust those random numbers to better represent the data.

'Start with random number -> look at data -> update random numbers -> look at data -> update random numbers
